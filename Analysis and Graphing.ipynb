```python
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
NC = {}
Gonz = {}
path = './DATA/2017/'
for dir_path , dir_name , file_names in os.walk(path+'North Carolina') :
    for name in file_names :
        if name[-3:] == 'adj' :
            player_name = name[:-4]
            NC[player_name] = pd.read_csv(os.path.join(dir_path,name)).drop('Unnamed: 0',axis=1)
for dir_path , dir_name , file_names in os.walk(path+'Gonzaga') :
    for name in file_names :
        if name[-3:] == 'adj' :
            player_name = name[:-4]
            Gonz[player_name] = pd.read_csv(os.path.join(dir_path,name)).drop('Unnamed: 0',axis=1)
```

```python
NC_two_stats = []
NC_players = []
for player,dataframe in NC.items() :
    stats = list(dataframe['2Pt %Att'])
    player_name = len(stats)*[player]
    for stat in stats :
        NC_two_stats.append(stat)
    for player in player_name :
        NC_players.append(player)
NC_twos = pd.DataFrame()
NC_twos['2Pt %Att'] = NC_two_stats
NC_twos['Player'] = NC_players
sns.set_style('whitegrid')
fig = plt.figure(figsize=(18,16))
ax = sns.violinplot(x=NC_twos['2Pt %Att'],y=NC_twos['Player'],data=NC_twos,order=NC.keys(),bw=.5,cut=.2)
plt.show()
```

```python
NC_three_stats = []
NC_players = []
for player,dataframe in NC.items() :
    stats = list(dataframe['3Pt %Att'])
    player_name = len(stats)*[player]
    for stat in stats :
        NC_three_stats.append(stat)
    for player in player_name :
        NC_players.append(player)
NC_threes = pd.DataFrame()
NC_threes['3Pt %Att'] = NC_three_stats
NC_threes['Player'] = NC_players
sns.set_style('whitegrid')
fig1 = plt.figure(figsize=(18,16))
ax1 = sns.violinplot(x=NC_threes['3Pt %Att'],y=NC_threes['Player'],data=NC_threes,order=NC.keys(),bw=.5,cut=.2)
plt.show()
```
