```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import scale
from sklearn.svm import SVC
from sklearn import random_projection
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from time import time
import os
```

## This is meant to test different machine learning algorithms on data 2

```python
current_year = 2017
important_years = [str(current_year-i) for i in range(4,0,-1)]
print(important_years)
train_data = []
'''
# One (Best ~71.29629% Random Forest, Depth 2 Num Estimators 60)
for year in important_years :
    if isinstance(train_data,list) :
        train_data = np.array(pd.read_csv('Data_2/New_Form_Data_'+year))
    else :
        train_data = np.vstack((train_data,np.array(pd.read_csv('Data_2/New_Form_Data_'+year))))
test_data = np.array(pd.read_csv('Data_2/New_Form_Data_'+str(current_year)))
train_x = train_data[:,:-1]
train_y = train_data[:,-1]
test = test_data[:,:-1]
test_act = test_data[:,-1]
'''
# Two (Best ~78.3882% Random Forest, Depth 7 Num Estimators 200)
# Next best is Random Forest, Depth 3 Num Estimatros 140
# Naive Bayes gives 75.36733%, PCA num components = 19
for year in important_years :
    if isinstance(train_data,list) :
        train_data = np.array(pd.read_csv('Data_2/New_Form_Combo_'+year))
    else :
        train_data = np.vstack((train_data,np.array(pd.read_csv('Data_2/New_Form_Combo_'+year))))
test_data = np.array(pd.read_csv('Data_2/New_Form_Combo_'+str(current_year)))
train_x = train_data[:,:-1]
train_y = train_data[:,-1]
train_x_scaled = StandardScaler().fit_transform(train_x)
test = test_data[:,:-1]
test_scaled = StandardScaler().fit_transform(test)
test_act = test_data[:,-1]
'''
# Three (Best ~75.06837, Naive Bayes, num components = 22)
for year in important_years :
    if isinstance(train_data,list) :
        train_data = np.array(pd.read_csv('Data_2/New_Form_Combo_'+year))
    else :
        train_data = np.vstack((train_data,np.array(pd.read_csv('Data_2/New_Form_Combo_'+year))))
test_data = np.array(pd.read_csv('Data_2/New_Form_Combo_'+str(current_year)))
train_x = train_data[:,:-1]
train_y = train_data[:,-1]
test = test_data[:,:-1]
test_act = test_data[:,-1]
'''
```

```python
C = [10**k for k in range(-10,11,1)]
accuracies = []
for c in C :
    classifier = LogisticRegression(C=1/c)
    classifier.fit(train_x,train_y)
    res = classifier.predict(test)
    acc = 1-np.count_nonzero([res[i]-test_act[i] for i in range(len(res))])/len(res)
    accuracies.append(acc)
accur = max(accuracies)
ind = accuracies.index(accur)
print('The best one was C={}, with an accuracy of {}.'.format(1/C[ind],accur))
```

```python
x = np.vstack((train_x,test))
y = np.hstack((train_y,test_act))
x_scaled = StandardScaler().fit_transform(x)
print(pd.DataFrame(data=x_scaled).describe())
```

```python
kernels = ['rbf','linear','poly','sigmoid']
degs = np.arange(1,6,1)
for ker in kernels :
    for degree in degs :
        svc = SVC(probability=True,kernel=ker,degree=degree)
        acc = cross_val_score(svc,x_scaled,y)
        print(f"Kernel : {ker}; Degree: {degree}; Acc: {acc}")
```

```python
pca_times = []
pca_scores = []
reg_times = []
reg_scores = []
for n in range(1,41) :
    pca = PCA(n_components=n)
    x_new = pca.fit_transform(x_scaled)
    pca_time = []
    reg_time = []
    pca_acc = []
    reg_acc = []

    gnb = GaussianNB()
    start = time()
    score = cross_val_score(gnb,x_new,y)
    end = time()
    pca_time.append(end-start)
    pca_acc.append(np.average(score))
    start = time()
    score = cross_val_score(gnb,x,y)
    end = time()
    reg_time.append(end-start)
    reg_acc.append(np.average(score))

    lgr = LogisticRegression(C=1e-8) # One C=0.01 Two C=0.01 Three C=1e-8
    start = time()
    score = cross_val_score(lgr,x_new,y)
    end = time()
    pca_time.append(end-start)
    pca_acc.append(np.average(score))
    start = time()
    score = cross_val_score(lgr,x,y)
    end = time()
    reg_time.append(end-start)
    reg_acc.append(np.average(score))

    svc = SVC(probability=True,kernel='sigmoid',degree=1)
    start = time()
    score = cross_val_score(svc,x_new,y)
    end = time()
    pca_time.append(end-start)
    pca_acc.append(np.average(score))
    start = time()
    score = cross_val_score(svc,x_scaled,y)
    end = time()
    reg_time.append(end-start)
    reg_acc.append(np.average(score))
    
    pca_times.append(pca_time)
    pca_scores.append(pca_acc)
    reg_times.append(reg_time)
    reg_scores.append(reg_acc)
```

```python
for n in range(1,41) :
    print("n={}         Naive Bayes              Log Reg.            SVM  ".format(n))
    print("Time -")
    print("     PCA : {}".format(pca_times[n-1]))
    print("     Reg : {}".format(reg_times[n-1]))
    print("Acc. -")
    print("     PCA : {}".format(pca_scores[n-1]))
    print("     Reg : {}".format(reg_scores[n-1]))
```

```python
combo = []
train_times = []
accs = []
# Experiment with max_depth and min_samples_leaf.
for max_depth in range(2,10) :
    for n_estimators in range(20,201,20) :
        combo.append((max_depth,n_estimators))
        clf = RandomForestClassifier(max_depth=max_depth,n_estimators=n_estimators)
        start = time()
        clf.fit(train_x,train_y)
        end = time()
        train_times.append(end-start)
        y_hat = clf.predict(test)
        accuracy = sum([int(test_act[i] == y_hat[i]) for i in range(len(y_hat))])/len(y_hat)
        accs.append(accuracy)
```

```python
print('Depth\tNum Estimators\tTime\t\t\tAccuracy')
for i in range(len(combo)) :
    print(str(combo[i][0])+'\t'+str(combo[i][1])+'\t\t'+str(train_times[i])+'\t'+str(accs[i]))
```

```python
gda = QDA()
start = time()
gda.fit(train_x,train_y)
y_hat = gda.predict(test)
g_score = sum([y_hat[i] == test_act[i] for i in range(len(y_hat))])/len(y_hat)
end = time()
g_time = end - start

lgr = LogisticRegression(C=1e-8) # One C=0.01 Two C=0.01 Three C=1e-8
start = time()
lgr.fit(train_x,train_y)
y_hat = lgr.predict(test)
l_score = sum([y_hat[i] == test_act[i] for i in range(len(y_hat))])/len(y_hat)
end = time()
l_time = end - start

nb = GaussianNB()
start = time()
nb.fit(train_x,train_y)
y_hat = nb.predict(test)
n_score = sum([y_hat[i] == test_act[i] for i in range(len(y_hat))])/len(y_hat)
end = time()
n_time = end - start
```

```python
print('\tTime\t\t\tScore')
print('QDA :\t'+str(g_time)+'\t'+str(g_score))
print('LGR :\t'+str(l_time)+'\t'+str(l_score))
print('NB  :\t'+str(n_time)+'\t'+str(n_score))
```
