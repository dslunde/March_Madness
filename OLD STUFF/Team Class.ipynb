```python
from bs4 import BeautifulSoup
import Player
import requests as req
```

```python
base_html = 'https://kenpom.com/'

email = r"darrenlund59@gmail\.com"
pword = r"qwerty12345"

def get_page(html) :
    """
    Gets the html from (page source) of a file from a given url
    """
    page = req.get(html,auth=(email,pword))
    return page.content

class Team(object) :
    """
    Creates a Team for the NCAA Tournament analysis
    """
    
    def __init__(self,team_name) :
        '''
        Takes in the team page url from kenpom.com
        '''
        html = base_html + 'team.php?team=' + team_name
        team_html = get_page(html)
        soup = BeautifulSoup(team_html, 'html.parser')
        print(soup.find_all('div'))
        players = soup.find_all('div',attrs={'id' : 'players'})
        print(players)
        self.players = []
        for player in players :
            player_html = base_html + player.find_all('a')[0]['href']
            player_name = player.find_all('a')[0].get_text()
            print(player_name)
            self.players.append(Player(player_html))
```
